```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import mlxtend.evaluate
from mlxtend.evaluate import bias_variance_decomp
```

```python
data = pd.read_csv("ТУТ  ПЕРВЫЙ ЦСВ")
```

```python
X = np.reshape(np.array(data.x), (-1, 1))

Y = np.reshape(np.array(data.y), (-1, 1))

a = plt.scatter(X,Y,3)
```

```python
#пункт 1
#ЗДЕСЬ ГДЕ-ТО ДОЛЖЕН БЫЛ ПОЯВИТЬСЯ ХОЛДАУТ, НО Я ЕГО ВОСПРИНЯЛ КАК РАЗБИВКУ НА TRAIN И TEST
X_iter = X
errs = []
errs2 = []
for m in range(2,50):
    X_iter = np.hstack((X_iter, X**m))  
    X_train, X_test, Y_train, Y_test = train_test_split(X_iter, Y, train_size=0.7, test_size=0.3, random_state=42)
    n = X_test.shape[0]
    n2 = X_train.shape[0]
    model = LinearRegression().fit(X_train, Y_train)  
    predictions = model.predict(X_train)
    predictions_test = model.predict(X_test)
    errs.append(np.sum((Y_train-predictions)**2)/n2)
    errs2.append(np.sum((Y_test-predictions_test)**2)/n)
   
    k = predictions.shape[0]
    k2 = predictions_test.shape[0]
    err = np.sum((Y_train - predictions)**2)/k
    err2 = np.sum((Y_test - predictions_test)**2)/k2

    #print(err, err2)
    plt.plot(errs)
```

```python
plt.plot(errs2)
```

```python
#пункт 4
#ТУТ ДОЛЖЕН БЫЛ БЫТЬ РАСЧЕТ ДИСПЕРСИИ И ОТКЛИКА

X_iter = X
errs = []
errs2 = []
for m in range(2,50):
    X_iter = np.hstack((X_iter, X**m))  
    X_train, X_test, Y_train, Y_test = train_test_split(X_iter, Y, train_size=0.7, test_size=0.3, random_state=42)
    n = X_test.shape[0]
    n2 = X_train.shape[0]
    model = LinearRegression().fit(X_train, Y_train)  
    predictions = model.predict(X_train)
    predictions_test = model.predict(X_test)
    errs.append(np.sum((Y_train-predictions)**2)/n2)
    errs2.append(np.sum((Y_test-predictions_test)**2)/n)
   
    k = predictions.shape[0]
    k2 = predictions_test.shape[0]
    err = np.sum((Y_train - predictions)**2)/k
    err2 = np.sum((Y_test - predictions_test)**2)/k2

    #print(err, err2)
    #plt.plot(errs)
    
    #X_train = X_train
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        model, X_train, Y_train, X_test, Y_test, 
        loss='mse',
        num_rounds=100,
        random_seed=46)

print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
